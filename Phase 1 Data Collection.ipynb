There is some missing data in the data 

Lets look at it

```python
data_collected.isna().sum()
```

It is observed that the missing data belongs to the Poverty and is specific to union territories

We will impute the missing values using Simple imputer

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values=np.NaN)
data_collected.Poverty2004=imputer.fit_transform(data_collected['Poverty2004'].values.reshape(-1,1))[:,0]
data_collected.Poverty2011=imputer.fit_transform(data_collected['Poverty2011'].values.reshape(-1,1))[:,0]
data_collected.PovertyDecrease=imputer.fit_transform(data_collected['PovertyDecrease'].values.reshape(-1,1))[:,0]
```

```python
data_collected
```

It is observed that the unemployment columns have % for rural unemployment and % for urban unemployment

By finding the average of rural and urban unemployment converting two columns to only one column Unemployment% 

```python
data_collected['Unemp%2004']=data_collected[['UnempRural 2004','UnempUrban2004']].mean(axis=1)
```

```python
data_collected['Unemp%2010']=data_collected[['UnempRural2010','UnempUrban2010']].mean(axis=1)
```

```python
data_collected= data_collected.drop(columns=['UnempRural 2004','UnempUrban2004','UnempRural2010','UnempUrban2010'])
```

```python
data_collected.to_csv('data_collected.csv',index=False)
```
